# Assigning the lpGBTs to the Slinks

Main boundary conditions:
- Done per 120-degree sector.
- Each 120-degree sector has 28 FPGAs.
- Each FPGA has either:
  - 12× 16G Slink outputs => 336 Slink outputs, or
  - 8× 25G Slink outputs => 224 Slink outputs.

One idea:
- Straddling ECONDs imply that an lpGBT pair is the quantum of the system.
- Single lpGBTs need to be paired:
  - Sort their rates and then take largest and smallest in pairs.
- Pairs of lpGBTs are then assigned to individual Slinks:
  - Max of 8 lpGBT pairs per Slink.

# A playground for optimization functions

This problem cannot be solved with linear-programming techniques.

Since the assignment of a lpGBT to a Slink is a boolean variable, we're in the domain of [mixed-integer non-linear programming (MINLP)](https://www.google.com/search?q=python+MINLP).

I found two python libraries that model such problems: `GEKKO` and `pyomo`.

```python
def split_dict_equally(input_dict, chunks=2):
    "Splits dict by keys. Returns a list of dictionaries."
    # prep with empty dicts
    return_list = [dict() for idx in range(chunks)]
    idx = 0
    for k,v in input_dict.items():
        return_list[idx][k] = v
        if idx < chunks-1:  # indexes start at 0
            idx += 1
        else:
            idx = 0
    return return_list
```

```python
from dataclasses import dataclass
@dataclass
class Slink:
    totalEvSize: float
    linkList: list

from operator import attrgetter
def assign_link(link, Slinks):
    Slinks.sort(key=attrgetter('totalEvSize'))
    target = Slinks[0]
    target.linkList.append(link[0])
    target.totalEvSize += link[1]
    return Slinks

Slinks = [Slink(totalEvSize=0, linkList=[]) for _ in range(nSlinks)]

#for link in sortedLinkRates.itertuples(name='Link', index=False):
#    Slinks = assign_link(link, Slinks)

lpGBTs={}
lpGBTs.update(doubleGBTRates_1)
lpGBTs.update(singleGBTRates_1)
lpGBTs={k: v for k, v in sorted(lpGBTs.items(), key=lambda item: item[1],reverse=True)}
for link in lpGBTs.items():
    Slinks = assign_link(link, Slinks)
```

```python
def Sort(sub_li): 
    l = len(sub_li) 
    for i in range(0, l): 
        for j in range(0, l-i-1): 
            if (sub_li[j][1] > sub_li[j + 1][1]): 
                tempo = sub_li[j] 
                sub_li[j]= sub_li[j + 1] 
                sub_li[j + 1]= tempo 
    return sub_li 
```

```python
def SwapModule(slinks, SlinkAve,L1,L2,L3):
    ss=[i for i in {k: v for k, v in sorted(slinks.items(), key=lambda item: item[1],reverse=True)}.items()]
    for i in ss:
        if len(i[1][1])!=1:
            slinkLow=i
            break
    slinkHigh=ss[-1]
    lp=Sort(slinkLow[1][1])[0][0]
 
    for ilist,var,arr in zip(L1,L2,L3):
        if str(lp) in ilist:
            var[str(lp),slinkLow[0]].value=0
            var[str(lp),slinkHigh[0]].value=1
        else:
            continue
```

```python
def SetNumberofLinks1(model, i):
    return sum([ model.GKSingles1[i, j] for j in Slink ]) == 1
    
def SetNumberofLinks2(model, i):
    return sum([ model.GKDoubles1[i, j] for j in Slink ]) == 1
    
def SetNumberofLinks3(model, i):
    return sum([ model.GKSlinks[i, j] for j in FPGA ]) == 1
    
model.linkCutS1 = pyo.Constraint(Single1, rule = SetNumberofLinks1)
model.linkCutD1 = pyo.Constraint(Double1, rule = SetNumberofLinks2)
model.linkCutSK = pyo.Constraint(Slink, rule = SetNumberofLinks3)
```

```python
def maxGBTs(model, j):
    return sum([ model.GKSingles1[i, j] for i in Single1 ]) + 2*sum([ model.GKDoubles1[i, j] for i in Double1 ]) <= maxGBTsPerSlink

def PerFPGA(model, j):   
    return sum([ model.GKSlinks[i, j] for i in Slink ]) == SlinkPerFPGA 

model.maxlpGBT = pyo.Constraint(Slink, rule = maxGBTs)
model.PerFPGA = pyo.Constraint(FPGA,  rule = PerFPGA)
```

```python
def RateCut(model, j):   
    return sum([ singleGBTRates_1[int(i)]*model.GKSingles1[i, j] for i in Single1 ]) + sum([ doubleGBTRates_1[int(i)]*model.GKDoubles1[i, j] for i in Double1 ]) <= maxSlinkRatePyomo 
        
model.ratelimit = pyo.Constraint(Slink, rule = RateCut)
```

def OBJrule(model):
    GKslinkRatesVariance=0

        GKslinkRatesVariance +=  (GKslinkRates-GKavSlinkRate)**2  
    GKslinkRatesStDev = pyo.sqrt(GKslinkRatesVariance / (nSlinks-1))
    return GKslinkRatesStDev

```python
def OBJrule(model):
    GKslinkRatesVariance=0
    for j in Slink:
        GKslinkRates=sum([ singleGBTRates_1[int(i)]*model.GKSingles1[i, j] for i in Single1 ]) + sum([ doubleGBTRates_1[int(i)]*model.GKDoubles1[i, j] for i in Double1 ]) 
        GKslinkRatesVariance +=  abs(GKslinkRates-GKavSlinkRate)  
    GKslinkRatesStDev = (GKslinkRatesVariance / (nSlinks-1))
    return GKslinkRatesStDev
model.objective = pyo.Objective(rule=OBJrule, sense=pyo.minimize)
```

```python
model.GKSingles1.fix()
model.GKDoubles1.fix()

model.GKSingles2 = pyo.Var(Single2, Slink, initialize = 0, within = pyo.Binary)
#model.GKDoubles2 = pyo.Var(Double2, Slink, initialize = 0, within = pyo.Binary)

def SetNumberofLinks12(model, i):
    return sum([ model.GKSingles2[i, j] for j in Slink ]) == 1
#def SetNumberofLinks22(model, i):
#    return sum([ model.GKDoubles2[i, j] for j in Slink ]) == 1
model.linkCutS2 = pyo.Constraint(Single2, rule = SetNumberofLinks12)
#model.linkCutD2 = pyo.Constraint(Double2, rule = SetNumberofLinks22)
    
model.del_component(model.maxlpGBT)
model.del_component(model.maxlpGBT_index)
def maxGBTs(model, j):
    #return sum([ model.GKSingles1[i, j].value for i in Single1]) + sum( [ model.GKSingles2[i, j] for i in Single2 ]) + 2*sum([ model.GKDoubles1[i, j].value  for i in Double1 ]) + 2*sum([model.GKDoubles2[i, j] for i in Double2 ]) <= maxGBTsPerSlink
    return sum([ model.GKSingles1[i, j].value for i in Single1]) + sum( [ model.GKSingles2[i, j] for i in Single2 ]) + 2*sum([ model.GKDoubles1[i, j].value  for i in Double1 ])  <= maxGBTsPerSlink
model.maxlpGBT = pyo.Constraint(Slink, rule = maxGBTs)

model.del_component(model.ratelimit)
model.del_component(model.ratelimit_index)
def RateCut(model, j):   
    #return sum([ singleGBTRates_1[int(i)]*model.GKSingles1[i, j].value for i in Single1])+sum( [singleGBTRates_2[int(i)]*model.GKSingles2[i, j] for i in Single2 ])  + sum([ doubleGBTRates_1[int(i)]*model.GKDoubles1[i, j].value for i in Double1])   +  sum( [doubleGBTRates_2[int(i)]*model.GKDoubles2[i, j] for i in Double2 ]) <= maxSlinkRatePyomo  
    return (1 , sum([ singleGBTRates_1[int(i)]*model.GKSingles1[i, j] for i in Single1])+sum( [singleGBTRates_2[int(i)]*model.GKSingles2[i, j] for i in Single2 ])  + sum([ doubleGBTRates_1[int(i)]*model.GKDoubles1[i, j] for i in Double1])  , maxSlinkRatePyomo ) 
model.ratelimit = pyo.Constraint(Slink, rule = RateCut)

model.del_component(model.objective)
def OBJrule(model):
    GKslinkRatesVariance=0
    for j in Slink:
        #GKslinkRates=sum([ singleGBTRates_1[int(i)]*model.GKSingles1[i, j].value for i in Single1])+sum( [singleGBTRates_2[int(i)]*model.GKSingles2[i, j] for i in Single2 ])  + sum([ doubleGBTRates_1[int(i)]*model.GKDoubles1[i, j].value for i in Double1])   +  sum( [doubleGBTRates_2[int(i)]*model.GKDoubles2[i, j] for i in Double2 ])
        GKslinkRates=sum([ singleGBTRates_1[int(i)]*model.GKSingles1[i, j] for i in Single1])+sum( [singleGBTRates_2[int(i)]*model.GKSingles2[i, j] for i in Single2 ])  + sum([ doubleGBTRates_1[int(i)]*model.GKDoubles1[i, j] for i in Double1]) 
        GKslinkRatesVariance +=  abs(GKslinkRates-GKavSlinkRate)  
    GKslinkRatesStDev = (GKslinkRatesVariance / (nSlinks-1))
    return GKslinkRatesStDev
model.objective = pyo.Objective(rule=OBJrule, sense=pyo.minimize)
```

```python
model.GKSingles2.fix()
#model.GKDoubles1.fix()

model.GKSingles3 = pyo.Var(Single3, Slink, initialize = 0, within = pyo.Binary)
#model.GKDoubles2 = pyo.Var(Double2, Slink, initialize = 0, within = pyo.Binary)

def SetNumberofLinks13(model, i):
    return sum([ model.GKSingles3[i, j] for j in Slink ]) == 1
#def SetNumberofLinks22(model, i):
#    return sum([ model.GKDoubles2[i, j] for j in Slink ]) == 1
model.linkCutS3 = pyo.Constraint(Single3, rule = SetNumberofLinks13)
#model.linkCutD3 = pyo.Constraint(Double2, rule = SetNumberofLinks22)
    
model.del_component(model.maxlpGBT)
model.del_component(model.maxlpGBT_index)
def maxGBTs(model, j):
    #return sum([ model.GKSingles1[i, j].value for i in Single1]) + sum( [ model.GKSingles2[i, j] for i in Single2 ]) + 2*sum([ model.GKDoubles1[i, j].value  for i in Double1 ]) + 2*sum([model.GKDoubles2[i, j] for i in Double2 ]) <= maxGBTsPerSlink
    return sum([ model.GKSingles1[i, j].value for i in Single1]) + sum( [ model.GKSingles2[i, j] for i in Single2 ]) + sum( [ model.GKSingles3[i, j] for i in Single3 ]) + 2*sum([ model.GKDoubles1[i, j].value  for i in Double1 ])  <= maxGBTsPerSlink
model.maxlpGBT = pyo.Constraint(Slink, rule = maxGBTs)

model.del_component(model.ratelimit)
model.del_component(model.ratelimit_index)
def RateCut(model, j):   
    #return sum([ singleGBTRates_1[int(i)]*model.GKSingles1[i, j].value for i in Single1])+sum( [singleGBTRates_2[int(i)]*model.GKSingles2[i, j] for i in Single2 ])  + sum([ doubleGBTRates_1[int(i)]*model.GKDoubles1[i, j].value for i in Double1])   +  sum( [doubleGBTRates_2[int(i)]*model.GKDoubles2[i, j] for i in Double2 ]) <= maxSlinkRatePyomo  
    return (1, sum([ singleGBTRates_1[int(i)]*model.GKSingles1[i, j] for i in Single1])+sum( [singleGBTRates_2[int(i)]*model.GKSingles2[i, j] for i in Single2 ])  +sum( [singleGBTRates_3[int(i)]*model.GKSingles3[i, j] for i in Single3 ]) + sum([ doubleGBTRates_1[int(i)]*model.GKDoubles1[i, j] for i in Double1])  , maxSlinkRatePyomo )  
model.ratelimit = pyo.Constraint(Slink, rule = RateCut)

model.del_component(model.objective)
def OBJrule(model):
    GKslinkRatesVariance=0
    for j in Slink:
        #GKslinkRates=sum([ singleGBTRates_1[int(i)]*model.GKSingles1[i, j].value for i in Single1])+sum( [singleGBTRates_2[int(i)]*model.GKSingles2[i, j] for i in Single2 ])  + sum([ doubleGBTRates_1[int(i)]*model.GKDoubles1[i, j].value for i in Double1])   +  sum( [doubleGBTRates_2[int(i)]*model.GKDoubles2[i, j] for i in Double2 ])
        GKslinkRates=sum([ singleGBTRates_1[int(i)]*model.GKSingles1[i, j] for i in Single1])+sum( [singleGBTRates_2[int(i)]*model.GKSingles2[i, j] for i in Single2 ]) + sum( [singleGBTRates_3[int(i)]*model.GKSingles3[i, j] for i in Single3 ]) + sum([ doubleGBTRates_1[int(i)]*model.GKDoubles1[i, j] for i in Double1]) 
        #GKslinkRatesVariance +=  abs(GKslinkRates-GKavSlinkRate)
        GKslinkRatesVariance +=  (GKslinkRates-GKavSlinkRate)**2
    #GKslinkRatesStDev = (GKslinkRatesVariance / (nSlinks-1))
    GKslinkRatesStDev = pyo.sqrt(GKslinkRatesVariance / (nSlinks-1))
    return GKslinkRatesStDev
model.objective = pyo.Objective(rule=OBJrule, sense=pyo.minimize)
```

```python
model.GKSingles3.fix()
#model.GKDoubles1.fix()

model.GKSingles4= pyo.Var(Single4, Slink, initialize = 0, within = pyo.Binary)
#model.GKDoubles2 = pyo.Var(Double2, Slink, initialize = 0, within = pyo.Binary)

def SetNumberofLinks14(model, i):
    return sum([ model.GKSingles4[i, j] for j in Slink ]) == 1
#def SetNumberofLinks22(model, i):
#    return sum([ model.GKDoubles2[i, j] for j in Slink ]) == 1
model.linkCutS4 = pyo.Constraint(Single4, rule = SetNumberofLinks14)
#model.linkCutD3 = pyo.Constraint(Double2, rule = SetNumberofLinks22)
    
model.del_component(model.maxlpGBT)
model.del_component(model.maxlpGBT_index)
def maxGBTs(model, j):
    #return sum([ model.GKSingles1[i, j].value for i in Single1]) + sum( [ model.GKSingles2[i, j] for i in Single2 ]) + 2*sum([ model.GKDoubles1[i, j].value  for i in Double1 ]) + 2*sum([model.GKDoubles2[i, j] for i in Double2 ]) <= maxGBTsPerSlink
    return sum([ model.GKSingles1[i, j].value for i in Single1]) + sum( [ model.GKSingles2[i, j] for i in Single2 ]) + sum( [ model.GKSingles3[i, j] for i in Single3 ])+ sum( [ model.GKSingles4[i, j] for i in Single4 ]) + 2*sum([ model.GKDoubles1[i, j].value  for i in Double1 ])  <= maxGBTsPerSlink
model.maxlpGBT = pyo.Constraint(Slink, rule = maxGBTs)

model.del_component(model.ratelimit)
model.del_component(model.ratelimit_index)
def RateCut(model, j):   
    #return sum([ singleGBTRates_1[int(i)]*model.GKSingles1[i, j].value for i in Single1])+sum( [singleGBTRates_2[int(i)]*model.GKSingles2[i, j] for i in Single2 ])  + sum([ doubleGBTRates_1[int(i)]*model.GKDoubles1[i, j].value for i in Double1])   +  sum( [doubleGBTRates_2[int(i)]*model.GKDoubles2[i, j] for i in Double2 ]) <= maxSlinkRatePyomo  
    return (1, sum([ singleGBTRates_1[int(i)]*model.GKSingles1[i, j] for i in Single1])+sum( [singleGBTRates_2[int(i)]*model.GKSingles2[i, j] for i in Single2 ])  +sum( [singleGBTRates_3[int(i)]*model.GKSingles3[i, j] for i in Single3 ])+sum( [singleGBTRates_4[int(i)]*model.GKSingles4[i, j] for i in Single4 ]) + sum([ doubleGBTRates_1[int(i)]*model.GKDoubles1[i, j] for i in Double1])  , maxSlinkRatePyomo )  
model.ratelimit = pyo.Constraint(Slink, rule = RateCut)

model.del_component(model.objective)
def OBJrule(model):
    GKslinkRatesVariance=0
    for j in Slink:
        #GKslinkRates=sum([ singleGBTRates_1[int(i)]*model.GKSingles1[i, j].value for i in Single1])+sum( [singleGBTRates_2[int(i)]*model.GKSingles2[i, j] for i in Single2 ])  + sum([ doubleGBTRates_1[int(i)]*model.GKDoubles1[i, j].value for i in Double1])   +  sum( [doubleGBTRates_2[int(i)]*model.GKDoubles2[i, j] for i in Double2 ])
        GKslinkRates=sum([ singleGBTRates_1[int(i)]*model.GKSingles1[i, j] for i in Single1])+sum( [singleGBTRates_2[int(i)]*model.GKSingles2[i, j] for i in Single2 ]) + sum( [singleGBTRates_3[int(i)]*model.GKSingles3[i, j] for i in Single3 ]) + sum( [singleGBTRates_4[int(i)]*model.GKSingles4[i, j] for i in Single4 ]) + sum([ doubleGBTRates_1[int(i)]*model.GKDoubles1[i, j] for i in Double1]) 
        GKslinkRatesVariance +=  abs(GKslinkRates-GKavSlinkRate)  
    GKslinkRatesStDev = (GKslinkRatesVariance / (nSlinks-1))
    return GKslinkRatesStDev
model.objective = pyo.Objective(rule=OBJrule, sense=pyo.minimize)
```
